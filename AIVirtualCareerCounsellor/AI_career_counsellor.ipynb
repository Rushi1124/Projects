```python
import nltk
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

# Categories & keywords
career_keywords = {
    "tech": ["coding", "programming", "ai", "machine learning", "software", "technology"],
    "arts": ["painting", "drawing", "design", "writing", "creativity", "art"],
    "commerce": ["business", "finance", "accounting", "marketing", "numbers", "commerce"],
    "medicine": ["doctor", "nurse", "health", "hospital", "medicine"],
    "law": ["law", "court", "advocate", "legal"],
    "sports": ["football", "cricket", "athlete", "sports", "gym"]
}

# Career suggestions
careers = {
    "tech": "Software Engineer, Data Scientist, AI/ML Engineer, Web Developer",
    "arts": "Graphic Designer, Content Writer, Fine Artist, Animator",
    "commerce": "Accountant, Business Analyst, Marketing Manager, Financial Analyst",
    "medicine": "Doctor, Nurse, Pharmacist, Physiotherapist",
    "law": "Lawyer, Judge, Legal Advisor, Paralegal",
    "sports": "Athlete, Coach, Sports Analyst, Physical Trainer"
}

# Main loop for multiple users
while True:
    user_input = input("\nTell me about your interests (or type 'exit' to quit): ").lower().split()
    if "exit" in user_input:
        print("Goodbye!")
        break

    user_words = [stemmer.stem(word) for word in user_input]
    suggestion = None
    for category, keywords in career_keywords.items():
        for word in user_words:
            if stemmer.stem(word) in [stemmer.stem(k) for k in keywords]:
                suggestion = category
                break
        if suggestion:
            break

    if suggestion:
        print(f"Suggested careers for you: {careers[suggestion]}")
    else:
        print("Sorry, we don't have suggestions for that interest yet.")
```
